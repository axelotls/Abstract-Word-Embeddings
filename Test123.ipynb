```python
#text = "Canada is a very beautiful country.It is the second largest country in the world."

import nltk
import gensim 
from nltk import sent_tokenize
from nltk import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
new_stopwords = ["the", "is", "a", "to", "in", "of", "with"]
stop.extend(new_stopwords)
print(stop)
```

```python
with open ('./polymer.txt') as fin:
    tokens = [word_tokenize(sent) for sent in fin]
tokens
```

```python
from gensim.models import Word2Vec
model = gensim.models.Word2Vec(tokens,
                               min_count=2,
                               vector_size=100,
                              workers=3,
                              window=5
                              )
```

```python
print(model)
```

```python
words = list(model.wv.index_to_key)
print(words)
```
